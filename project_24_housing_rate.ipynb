# One Hot Encoder(dummie)

```python
db_cat=db['ocean_proximity']
db_cat = pd.get_dummies(db_cat)
db_cat.head()
```

# standard scaler

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

scalar.fit(db.drop(['ocean_proximity','median_house_value'],axis=1))
scale_features = scalar.transform(db.drop(['ocean_proximity','median_house_value'],axis=1)

df_feat = pd.DataFrame(scale_features,columns = db.columns[:-1])
df_feat.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
db_stand = db.drop(['ocean_proximity','median_house_value'],axis=1)
scalar.fit(db_stand)
scale_features = scalar.transform(db_stand)
df_feat = pd.DataFrame(scale_features,columns = db_stand.columns[:])
df_feat.head()
```

```python
db_final = pd.concat([df_feat, db_cat,db.median_house_value], axis=1)
db_final.head()
```

# Modeling

```python
X=db_final.drop('median_house_value',1)
X['total_rooms'] =     np.cbrt(X.total_rooms)
X['total_bedrooms'] =  np.cbrt(X.total_bedrooms)
X['population'] =      np.cbrt(X.population)
X['households'] =      np.cbrt(X.households)
X['median_income'] =   np.cbrt(X.median_income)
X.rename(columns={'<1H OCEAN' : 'OCEAN'})
y=db_final['median_house_value']
```

```python
X.rename(columns={'<1H OCEAN' : 'OCEAN'},inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
# fit a model using linear model method from sklearn
from sklearn import linear_model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
```

```python
# generate predictions
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(X_train)
lin_mse = mean_squared_error(y_test,predictions)
lin_rmse = np.sqrt(lin_mse)
print("RMSE:", lin_rmse)
```

## find best parametrers using gridsearhCV and also check and remove overfitting and underfitting problem by using hyperparameter tuning

```python
def Snippet(model, X_test, y_test,X_train,y_train,parameters): 
    
    print(format('How to find parameters using GridSearchCV  for Regression','*^82'))    
    
    import warnings
    warnings.filterwarnings("ignore")
    
    # load libraries
    #from sklearn import datasets
    #from sklearn.model_selection import train_test_split
    from sklearn.model_selection import GridSearchCV
    #print(model,parameters)
    
    grid = GridSearchCV(estimator=model, param_grid = parameters, cv = 2,refit=True, n_jobs=-1)
    grid.fit(X_train, y_train)   
    
    y_pred = grid.best_estimator_.predict(X_test)
    train_y_pred = grid.best_estimator_.predict(X_train)
    test_RMSE = mean_squared_error(y_test, y_pred, squared=False)
    train_RMSE = mean_squared_error(y_train, train_y_pred, squared=False)
    
    fig, ax = plt.subplots()
    ax.scatter(y_test, y_pred)
    ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
    ax.set_xlabel('Measured')
    ax.set_ylabel('Predicted')
    
    
    # Results from Grid Search
    print("\n========================================================")
    print(" Results from Grid Search " )
    print("========================================================")    
    
    print("\n The best estimator across ALL searched params:\n",
          grid.best_estimator_)
    
    print("\n The best score across ALL searched params:\n",
          grid.best_score_)
    
    print("\n The best parameters across ALL searched params:\n",
          grid.best_params_)
                                    
    print("\n The train_RMSE across ALL models:\n",
          train_RMSE)
                                    
    print("\n The test_RMSE across ALL models:\n",
          test_RMSE) 
    
    print (plt.show())
    
    print("\n ========================================================")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
alg = [XGBRegressor(), GradientBoostingRegressor(),RandomForestRegressor()]
parameters = [{'learning_rate': [0.01,0.02,0.03],
                  'subsample'    : [0.9, 0.5, 0.2],
                  'n_estimators' : [100,500,1000],
                  'max_depth'    : [4,6,8]},
                
              {'learning_rate': [0.01,0.02,0.03],
                  'subsample'    : [0.9, 0.5, 0.2],
                  'n_estimators' : [100,500,1000],
                  'max_depth'    : [4,6,8]},
              {'max_depth'    : [4,6,8],
              'n_estimators' : [100,500,1000],
              'min_samples_leaf':[1,2,3],
              'random_state': [40,45,50]}
             ]


for i in range(len(alg)):
    #print(alg[i])
                              
    Snippet(alg[i] ,X_test, y_test,X_train,y_train,parameters[i])
```
